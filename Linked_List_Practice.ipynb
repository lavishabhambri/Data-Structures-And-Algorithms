# Eliminate duplicates from LL

```python
# Given a sorted linked list (elements are sorted in ascending order). Eliminate duplicates from the given LL, such that output LL contains only unique elements.
#You don't need to print the elements, just remove duplicates and return the head of updated LL.
#Input format : Linked list elements (separated by space and terminated by -1)


class node:
    
    def __init__(self, data):
        self.data = data
        self.next = None

#to take input of the linked list
def taking_input():
    
    inputList =  [int(element) for element in input().split()]
    head = None
    tail = None
    for currentData in  inputList:
        if currentData == -1:
            break
            
        newNode = node(currentData)
        if head is None:
            head  = newNode
            tail = newNode
            
        else:
            tail.next = newNode
            tail = newNode
            
    return head



#to print the elements of linked list
def printLL(head):
    
    while head is not None:
        print(str(head.data) + "->" ,end="")
        head = head.next
    print("None")
    return

#to get the length of LL
def length(head):
    count = 0
    while head is not None:
        count = count + 1
        head = head.next
    return count


#to remove the duplicates from the sorted LL
def removeDuplicates(head):
    
    previous = head
    current = head.next
    
    if previous == None or current == None:
        return head
    
    while current is not None:
        if previous.data is not current.data:
            previous.next = current
            previous = current
            current = current.next
        else:
            current = current.next
        
    previous.next = current
        
    return head

        
    
    
head = taking_input()
printLL(head)
head = removeDuplicates(head)
printLL(head)
#removeDuplicates(head)
```

# Midpoint of LL

# Merge Sort LL

```python
class node:
    
    def __init__(self, data):
        self.data = data
        self.next = None
        
def taking_input():
    inputList = [int(element) for element in input().split()]
    
    head = None
    tail = None
    for currentData in inputList:
        if currentData == -1:
            break
            
        newNode = node(currentData)
        if head is None:
            head = newNode
            tail = newNode
            
        else:
            tail.next = newNode
            tail = newNode
            
    return head

def printLL(head):
    
    while head is not None:
        print(str(head.data) + "->" ,end="")
        head = head.next
    print("None")
    return


#to merge two sorted Linked lists
def merge(h1, h2):
    
    fh = None
    ft = None
    
    if h1.data > h2.data:
        fh = h2
        h2 = h2.next
    
    else:
        fh = h1
        h1 = h1.next
        
    ft = fh   
    while (h1 != None and h2 != None):
        
        if (h1.data > h2.data):
            ft.next = h2
            ft = ft.next
            h2 = h2.next
            
        elif (h1.data< h2.data):
            ft.next = h1
            ft = ft.next
            h1 = h1.next
            
    if (h1 is not None):
        ft.next = h1
    
    elif (h2 is not None):
        ft.next = h2
        
    return fh

#to find the midpoint of Linked list
def midpoint(head):
    
    slow = head
    fast = head
    while (fast.next != None and (fast.next).next != None):
        fast = (fast.next).next
        slow = slow.next
        
    return slow


#applying mergesort on both LL
def mergeSort(head):
    
    if head.next == None:
        return head
    
    mid = midpoint(head)
    head1 = head
    head2 = mid.next
    mid.next = None
    
    h1 = mergeSort(head1)
    h2 = mergeSort(head2)
    
    return merge(h1, h2)
    

head = taking_input()
printLL(head)
head = mergeSort(head)
printLL(head)
```

# Find a node in LL (recursive)

```python
# Given a linked list and an integer n you need to find and return index where n is present in the LL. Do this recursively.
#Return -1 if n is not present in the LL.
#Indexing of nodes starts from 0.

class node:
    
    def __init__(self, data):
        self.data = data
        self.next = None

#to take input of the linked list
def taking_input():
    
    inputList =  [int(element) for element in input().split()]
    head = None
    tail = None
    for currentData in  inputList:
        if currentData == -1:
            break
            
        newNode = node(currentData)
        if head is None:
            head  = newNode
            tail = newNode
            
        else:
            tail.next = newNode
            tail = newNode
            
    return head



#to print the elements of linked list
def printLL(head):
    
    while head is not None:
        print(str(head.data) + "->" ,end="")
        head = head.next
    print("None")
    return


#to find a node recursively in the LL
def recursive(head, n):
    
    if head == None:
        return None
    count = 1
    if head.data == n:
        return count
    
    count = recursive(head.next, n) + count
    return count 
    
head = taking_input()
printLL(head)
head = recursive(head, 3)
print(head)
```

# Delete every N nodes

```python
#Given a linked list and two integers M and N. Traverse the linked list such that you retain M nodes then delete next N nodes, continue the same until end of the linked list. That is, in the given linked list you need to delete N nodes after every M nodes.
#Input format :
#Line 1 : Linked list elements (separated by space and terminated by -1)
#Line 2 : M
#Line 3 : N

class node:
    
    def __init__(self, data):
        self.data = data
        self.next = None

#to take input of the linked list
def taking_input():
    
    inputList =  [int(element) for element in input().split()]
    head = None
    tail = None
    for currentData in  inputList:
        if currentData == -1:
            break
            
        newNode = node(currentData)
        if head is None:
            head  = newNode
            tail = newNode
            
        else:
            tail.next = newNode
            tail = newNode
            
    return head



#to print the elements of linked list
def printLL(head):
    
    while head is not None:
        print(str(head.data) + "->" ,end="")
        head = head.next
    print("None")
    return


#to get the length of LL
def length(head):
    count = 0
    while head is not None:
        count = count + 1
        head = head.next
    return count

#to delete m elements after every n elements
def delete(head, m , n):
    
    if m == 0:
        return None
    
    if n == 0:
        return head
    
    currentNode = head
    temp = None
    
    while currentNode != None:
        take = 0
        skip = 0
        
        while currentNode != None and take < m:
            if temp == None:
                temp = currentNode
            else:
                temp.next = currentNode
                temp = currentNode
                
            currentNode = currentNode.next
            take += 1
        while currentNode != None and skip < n:
            newNode = currentNode
            currentNode = newNode.next
            skip += 1
            
    if temp != None:
        temp.next = None
        
    return head
                
head = taking_input()
printLL(head)
head = delete(head, 3, 2)
printLL(head)
```

# Swap two Node of LL

```python
#Given a linked list, i & j, swap the nodes that are present at i & j position in the LL. You need to swap the entire nodes, not just the data.
#Indexing starts from 0. You don't need to print the elements, just swap and return the head of updated LL.
#Assume i & j given will be within limits. And i can be greater than j also.
#Input format :
#Line 1 : Linked list elements (separated by space and terminated by -1)
#Line 2 : i and j (separated by space)


class node:
    
    def __init__(self, data):
        self.data = data
        self.next = None

#to take input of the linked list
def taking_input():
    
    inputList =  [int(element) for element in input().split()]
    head = None
    tail = None
    for currentData in  inputList:
        if currentData == -1:
            break
            
        newNode = node(currentData)
        if head is None:
            head  = newNode
            tail = newNode
            
        else:
            tail.next = newNode
            tail = newNode
            
    return head



#to print the elements of linked list
def printLL(head):
    
    while head is not None:
        print(str(head.data) + "->" ,end="")
        head = head.next
    print("None")
    return


#to get the length of LL
def length(head):
    count = 0
    while head is not None:
        count = count + 1
        head = head.next
    return count


#to swap the two elements in LL
def swap(head, i, j):
    
    p1 = None
    p2 = None
    c1 = head
    c2 = head
    count1 = 0
    count2 = 0
    while c1 != None:
        if c1 != None and count1 != i:
            p1 = c1
            c1 = c1.next
            count1 = count1 + 1
        else:
            break
     
    while c2 != None:
        if c1 != None and count2 != j:
            p2 = c2
            c2 = c2.next
            count2 = count2 + 1
        else:
            break
    while c1 != None and c2 != None:        
        if i == 0 or j == 0:
            p2.next = c1
            c2.next = c1.next
            c1.next = c2.next
        elif abs(i-j) == 1:
            p1.next = c2
            p2.next = c1
            c1.next = c2.next
        elif abs(i-j) == 1 and i == 0 or j == 0:
            c2.next = c1
            c1.next = c2.next
            p2.next = c1
        else:
            p1.next = c2
            p2.next = c1
            c2.next = c1.next
            c1.next = c2.next
        
    return head

head = taking_input()
printLL(head)
head = swap(head, 2,3)
printLL(head)
            
```

# kReverse

```python
#Implement kReverse( int k ) in a linked list i.e. you need to reverse first K elements then reverse next k elements and join the linked list and so on.
#Indexing starts from 0. If less than k elements left in the last, you need to reverse them as well. If k is greater than length of LL, reverse the complete LL.
#You don't need to print the elements, just return the head of updated LL.


class node:
    
    def __init__(self, data):
        self.data = data
        self.next = None

#to take input of the linked list
def taking_input():
    
    inputList =  [int(element) for element in input().split()]
    head = None
    tail = None
    for currentData in  inputList:
        if currentData == -1:
            break
            
        newNode = node(currentData)
        if head is None:
            head  = newNode
            tail = newNode
            
        else:
            tail.next = newNode
            tail = newNode
            
    return head



#to print the elements of linked list
def printLL(head):
    
    while head is not None:
        print(str(head.data) + "->" ,end="")
        head = head.next
    print("None")
    return


#to get the length of LL
def length(head):
    count = 0
    while head is not None:
        count = count + 1
        head = head.next
    return count


#to reverse the LL
def reverse(head):
    
    if head == None:
        return None
    
    if head.next == None:
        return head
    
    smallHead = reverse(head.next)
    tail = head.next
    tail.next = head
    head.next = None
    return smallHead, head

#to reverse k elements
def kReverse(head, k):
    
    count = 1
    h1 = head
    t1 = head
    if length(head) < k:
        return reverse(head)
    
    while count < k:
        t1 = t1.next
        count = count + 1
        
    h2 = t1.next
    t1.next = None
    h1 , t1 = reverse(h1)
    h2 = kReverse(h2, k)
    t1.next = h2
    
    return h1

head = taking_input()
printLL(head)
head = kReverse(head, 3)
printLL(head)
    
```
